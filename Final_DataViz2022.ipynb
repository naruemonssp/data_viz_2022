# **ข้อมูลชุดที่ 1 การตรวจโรค (ov)**

```python
data1 =  pd.read_csv(os.path.join(path,'alldb.csv')) # ทำการนำเข้าข้อมูล การตรวจโรค(ov) โดยการเรียกใช้ข้อมูลประเภท .csv ใน drive ไฟล์ที่เรียกใช้คือ alldb.csv  โดยตั้งชื่อตารางว่า data1
data1.head() # print data ดูว่ามันหน้าตาเป็นยังไง
```

```python
data1.shape  #ตรวจสอบขนาดของข้อมูล
```

```python
data1.dtypes  #ตรวจสอบชนิดของข้อมูล
```

## **แสดง Columns ที่สนใจ**

```python
data = data1[['user','RowKey','date','ov','phone','site','time','urine','PartitionKey','title','age']]
data
```

## **ตรวจสอบค่า Missing**

```python
data.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง
```

```python
data_drop = data.dropna() # แก้ไขค่า missing โดยทำการ drop ค่า missing ออกไปโดยใช้คำสั่ง dropna()
data_drop
```

## **เปลี่ยนชนิดข้อมูลให้เป็นไปตามข้อกำหนด**

### user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data_drop = data_drop[(data_drop['user'].astype(str).str.isdigit()) & (data_drop['user'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ user ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
print(data_drop.shape)                                                                             # .str.isdigit ใช้ตรวจสอบว่าอักขระทั้งหมดเป็นตัวเลขหรือไม่
data_drop.head()
```

### RowKey เป็น String

```python
import numpy as np
```

```python
data_drop['RowKey'] = data_drop['RowKey'].astype('str')
print(data_drop.shape)
data_drop.head()
```

### เปลี่ยน date เป็น YYYY-MM-DD

```python
import datetime # ทำการ import package datetime 
data_drop = pd.DataFrame(data_drop) 
data_drop['date'] = pd.to_datetime(data_drop['date']) # format = "%d/%m/%Y" 
print(data_drop.shape)
data_drop.head()
```

### เปลี่ยน phone เป็น String

```python
data_drop['phone'] = data_drop['phone'].astype('str')
print(data_drop.shape)
data_drop.head()
```

### เปลี่ยน site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data_drop = data_drop[(data_drop['site'].astype(str).str.isdigit()) & (data_drop['site'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ site ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
print(data_drop.shape)                                                                            
data_drop.head()
```

### เปลี่ยน time เป็น hh-mm-ss

```python
data_drop['time'] = data_drop['time'].replace(':','-',regex=True) # กำหนดรูปแบบของ time เป็น hh-mm-ss
data_drop
```

### เปลี่ยน urine มีค่า String

```python
data_drop['urine'] = data_drop['urine'].astype('str')
print(data_drop.shape)
data_drop.head()
```

### เปลี่ยน ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
set(data_drop['ov'])
```

```python
data_drop['ov'] = pd.to_numeric(data_drop['ov'], errors='coerce') # เปลี่ยนให้เป็นตัวเลขเท่านั้น
data_drop = data_drop.replace(np.nan, 0, regex=True) 
data_drop['ov'] = data_drop['ov'].astype(int)
```

```python
set(data_drop['ov'])
```

### เปลี่ยน PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น))

```python
data_drop = data_drop[(data_drop['PartitionKey'].astype(str).str.isdigit()) & (data_drop['PartitionKey'].str.len() == 5)]
print(data_drop.shape)
data_drop.head()
```

```python
data_drop.dtypes  #ตรวจสอบชนิดของข้อมูล
```

```python
data_drop = data_drop.replace(np.nan, 0, regex=True)
data_drop['user'] = data_drop['user'].astype(int)
data_drop['age'] = data_drop['age'].astype(int)
```

```python
data_drop.dtypes  #ตรวจสอบชนิดของข้อมูล
```

# **ข้อมูลชุดที่ 2 hospital list**

```python
data2  =  pd.read_excel(os.path.join(path,'hospital_list.csv')) # ทำการนำเข้าข้อมูล  hospital list โดยการเรียกใช้ข้อมูลประเภท .xlsx ใน drive ไฟล์ที่เรียกใช้คือ hospital_list.xlsx  โดยตั้งชื่อตารางว่า data2
data2.head() # print data ดูว่ามันหน้าตาเป็นยังไง
```

```python
data2.shape # ตรวจสอบขนาดของข้อมูล
```

## **ตรวจสอบค่า Missing**

```python
data2.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง
```

## **เปลี่ยนชนิดข้อมูลให้เป็นไปตามข้อกำหนด**

```python
data2 = data2[['hcode','name','zone_name','province','amphur','tambon','name5','short_english']]      # เลือกเฉพาะคอลัมน์ที่ต้องการ
print(data2.shape)
data2.head()
```

```python
data2['name'] = data2['name'].fillna('ไม่ระบุ') 
data2['zone_name'] = data2['zone_name'].fillna('ไม่ระบุ') 
data2['amphur'] = data2['amphur'].fillna('ไม่ระบุ') 
data2['tambon'] = data2['tambon'].fillna('ไม่ระบุ') 
data2['name5'] = data2['name5'].fillna('ไม่ระบุ') 
data2['short_english'] = data2['short_english'].fillna('ไม่ระบุ');
```

```python
data2.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง
```

# **รวมตารางข้อมูลทั้งหมด**

```python
data_all = data_drop.merge(data2,how='left',left_on='site',right_on='hcode')     # เชื่อมตาราง
data_all.head()
```

## **ตรวจสอบค่า Missing**

```python
data_all.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง
```
